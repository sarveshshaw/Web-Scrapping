```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
Links=[]
k=0
#till 72
for i in range(1,72):
    url="https://gympp.com/mumbai/"+str(i)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101Firefox/50.0'}
    source=requests.get(url, headers=headers)
    soup=bs(source.content,'html.parser')
    main=soup.find('div',class_='center_column')
    results=main.find_all('div',class_='mdl-card__supporting-text content center_column')
    for result in results:
        p1=result.find('div',class_='pname')
        if p1.a!=None:
            link=p1.a['href']
            k=k+1
            Links.append("https://gympp.com"+str(link))
```

```python
Name=[]
Add=[]
Phone1=[]
Phone2=[]
for link in Links:
    subheaders = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101Firefox/50.0'}
    subsource=requests.get(link, headers=headers)
    subsoup=bs(subsource.content,'html.parser')
    subresult=subsoup.find('ul',class_="product-info-list")
    subresult2=subsoup.find('h1',class_="name")
    if subresult2!=None :
        Name.append(subresult2.text.strip())
    else:
        Name.append("NaN")
    add=subresult.find('div',class_='goodtail')
    if add!=None:
        add.a.decompose()
        Add.append(add.text.strip())
    else:
        Add.append("NaN")
    num1=subresult.find_all('span',class_='number')[0]
    if num1!=None:
        Phone1.append(num1.text.strip())
    else:
        Phone1.append("NaN")
    num2=subresult.find_all('span',class_='number')[1]
    if num2!=None:
        Phone2.append(num2.text.strip())
    else:
        Phone2.append("NaN")
```

```python
df['Name']=Name
df['Address']=Add
df['Phone 1']=Phone1
df['Phone 2']=Phone2
df['Links']=Links
```
