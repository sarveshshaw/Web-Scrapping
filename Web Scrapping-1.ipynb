```python
import requests
```

```python
URL = 'https://www.monster.com/jobs/search/?q=Software-Developer&where=Australia'
page=requests.get(URL)
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
results=soup.find(id='ResultsContainer')
```

```python
print(results.prettify())
```

```python
jobs=results.find_all( 'section',class_='card-content')
Job=[]
Company=[]
Location=[]
Link=[]
```

```python
for job in jobs:
    title=job.find('h2',class_='title')
    company=job.find('div',class_='company')
    location=job.find('div',class_='location')
    link=job.find('a')
    if None in (title,company,location):
        continue
    print(title.text.strip())
    Job.append(title.text.strip())
    print(company.text.strip())
    Company.append(company.text.strip())
    print(location.text.strip())
    Location.append(location.text.strip())
    print("Link=",link['href'])
    Link.append(link['href'])
    print()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=['Job','Company','Location','Application Link'])
df['Job']=Job
df['Company']=Company
df['Location']=Location
df['Application Link']=Link
```
