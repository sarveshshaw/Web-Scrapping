```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.medifee.com/hospitals-in-new-delhi"
page=requests.get(url)
```

```python
soup=bs(page.content,'html.parser')
```

```python
soup
```

```python
content=soup.find('table',class_='table table-bordered table-striped')
```

```python
content
```

```python
table=content.find('tbody')
table
```

```python
results=table.find_all('tr')
results
```

```python
for result in results:
    print(result.td)
    print()
```

```python
Links=[]
for result in results:
    link=result.td.find('a')
    if link != None :
        print(link['href'])
        Links.append(link['href'])
        print()
```

```python
Name=[]
Add=[]
Phone=[]
for link in Links:
    suburl="https://www.medifee.com"+str(link)
    subpage=requests.get(suburl)
    subsoup=bs(subpage.content,'html.parser')
    info=subsoup.find('div',class_='col-lg-6')
    if info!= None:  
        name=info.h4
        Name.append(name.text)
        minfo=info.find_all('p')
        add=info.find_all('p')[0]
        if add!=None:
            Add.append(add.text)
        phone=info.find_all('p')[1]
        if phone!=None:
            for span in phone.span:
                span.replace_with('')
            number=phone.text
            no=number.split(':')
            Phone.append(no[1])
```

```python
import pandas as pd
df=pd.DataFrame(columns=['Name','Address','Phone'])
```

```python
df['Name']=Name
df['Address']=Add
df['Phone']=Phone
```
