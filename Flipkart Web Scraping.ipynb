```python
Name=[]
Price=[]
Rating=[]
Description=[]
```

```python
k=0
for i in range(1,25):
    url='https://www.flipkart.com/search?q=harry+potter&page='+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    results=soup.find_all('div',class_='_4ddWXP')
    for result in results:
        title=result.find('a',class_='s1Q9rs')
        desc=result.find('div',class_='_3Djpdu')
        rating=result.find('span',class_='_2_R_DZ')
        price=result.find('div',class_='_30jeq3')
        if None in (title,desc,rating,price):
            continue
        print(title['title'])
        Name.append(title['title'])
        print(desc.text)
        Description.append(desc.text)
        print(price.text)
        Price.append(price.text)
        print(rating.text)
        Rating.append(rating.text)
        print(title['href'][1:])
        k=k+1
        print()
```
