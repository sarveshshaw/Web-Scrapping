```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
Links=[]
#till 57
for i in range(1,57):
    url='https://www.fundoodata.com/citiesindustry/2/5/list-of-bfsi-companies-in-mumbai?&pageno='+str(i)+'&tot_rows=1111&total_results=1111&no_of_offices=0'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101Firefox/50.0'}
    source=requests.get(url, headers=headers)
    source.content
    soup=bs(source.content,'html.parser')
    results=soup.find_all('div',class_='search-result')
    for result in results:
        link=result.find('div',class_='heading')
        if link!=None:    
            Links.append(link.a['href'])
```

```python
Name=[]
Add=[]
Phone=[]
for link in Links:
    page=requests.get(link)
    page.content
    subsoup=bs(page.content,'html.parser')
    subresult=subsoup.find('div',class_='search-page-right-pannel')
    if subresult!=None:
        name=subresult.find('div',class_='search-page-heading-red')
        Name.append(name.text.strip())
        no=subresult.find('div',class_='detail-line')
        if no!=None:
            ph=no.text.strip()
            no=ph.split("\n")
            number=no[0]
            Phone.append(number)
        else:
            Phone.append("NaN")
        tr=subresult.find('div',class_='search-page-heading-red').decompose()
        add=subresult.text.strip().split("\t")
        Add.append(add[0])
```
